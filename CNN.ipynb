```python
import torch
import torch.nn as nn
conv = nn.Conv2d(
    in_channels=3,
    out_channels=8,
    kernel_size=3
)
relu = nn.ReLU()
x = torch.randn(1,3,32,32)
y = conv(x)
print("Input Shape :",x.shape)
print("Output Shape :",y.shape)
```

```python
conv = nn.Conv2d(
    in_channels=3,
    out_channels=8,
    kernel_size=3
)
relu = nn.ReLU()
pool = nn.MaxPool2d(kernel_size=2, stride=2)

x = torch.randn(1,3,32,32)
y = relu(conv(x))
z = pool(y)
print(y.shape)
print(z.shape)
```

```python
flat = z.view(z.size(0), -1)
print(flat.shape)
```

```python
import torch
import torch.nn as nn

conv1 = nn.Conv2d(3, 8, 3)      # 32 -> 30
pool  = nn.MaxPool2d(2, 2)      # 30 -> 15
conv2 = nn.Conv2d(8, 16, 3)     # 15 -> 13
pool2 = nn.MaxPool2d(2, 2)      # 13 -> 6

x = torch.randn(1, 3, 32, 32)
out = pool2(conv2(pool(conv1(x))))

print("out:", out.shape)  # should be (1,16,6,6)

flat = out.view(out.size(0), -1)
print("flat:", flat.shape) # (1,576)

fc = nn.Linear(576, 10)
logits = fc(flat)
print("logits:", logits.shape)  # (1,10)
```
